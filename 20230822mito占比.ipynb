```python
        out_file = open("/SGRNJ06/randd/USER/cjj/celedev/rna/20230821_10x/TS124N_mito.txt", 'w')
        MITO_GENE_PERCENT_LIST = [5, 10, 15, 20, 50]
        mt_pct_var = f'pct_counts_{MITO_VAR}'
        total_cell_number = adata.n_obs

        for mito_gene_percent in MITO_GENE_PERCENT_LIST:
            cell_number = sum(adata.obs[mt_pct_var] > mito_gene_percent)
            fraction = round(cell_number / total_cell_number * 100, 2)
            out_file.write(f'Fraction of cells have mito gene percent>{mito_gene_percent}% : {fraction}\n')
```

```python
out_file.close()
```

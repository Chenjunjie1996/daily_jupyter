```r
rds1 = AverageExpression(rds, group.by='sample')
```

```r
head(rds1$RNA)
```

```r
PRO1 <- CreateSeuratObject(counts = rds1$RNA, project = "NEW", min.cells = 0, min.features = 0)
PRO1@meta.data$sample <- rownames(PRO1@meta.data)
PRO1 <-  NormalizeData(object = PRO1)
PRO <-  ScaleData(object = PRO1)
```

```r
PRO@meta.data
```

```r
PRO =  FindVariableFeatures(PRO, selection.method = "vst", nfeatures = 2000)
```

```r
PRO <- RunPCA(object=PRO,npcs = 8, features = VariableFeatures(object = PRO))
```

```r
PRO <- FindNeighbors(PRO, reduction = "pca", dims = 1:20)
PRO <- FindClusters(PRO,resolution = 0.6, algorithm = 1)
PRO <- RunTSNE(object=PRO,dims.use=1:20,do.fast=TRUE,check_duplicates = FALSE)
PRO <- RunUMAP(PRO, reduction = "pca", dims = 1:20)
```

```r
DimPlot(PRO, reduction = "pca",group.by="sample",pt.size = 3) + scale_color_npg()
```

```r
import pandas as pd
from pandas import Series
import glob
```

```r
def calc_corr(a,b):
    s1 = Series(a)
    s2 = Series(b)
    return round(s1.corr(s2), 3)
```

```r
files = glob.glob("/SGRNJ06/randd/USER/cjj/celedev/rna/20231128nanda/*.tsv")
```

```r
files
```

```r
df = pd.read_csv("/SGRNJ06/randd/USER/cjj/celedev/rna/20231128nanda/Tcells.tsv", sep='\t')
```

```r
samples = list(df.columns)[1:]
```

```r
df_res = pd.DataFrame()
for sample1 in samples:
    temp_list = []
    for sample2 in samples:
        temp_list.append(calc_corr(df[sample1], df[sample2]))
    df_res[sample1] = temp_list
        
```

```r
df_res.index = samples
```

```r
df_res
```

```r
df_res.to_csv("/SGRNJ06/randd/USER/cjj/celedev/rna/20231128nanda/T_corr.tsv", sep='\t', index=True)
```

```python
import pandas as pd 
import numbers
```

```python
df = pd.read_csv("/SGRNJ06/randd/PROJECT/RD20073101_ScRNA_VDJ/bulk_TCR/20230703_limit2/new_pipeline/frozen_1w/Mus_0626Spleen_freeze_1w_T3lib/04.mapping_vdj/Mus_0626Spleen_freeze_1w_T3lib_airr.tsv", sep='\t')
```

```python
df.fillna("", inplace=True)
```

```python
df["index"] = df["sequence_id"].apply(lambda x: x.split('_')[0])
```

```python
df
```

```python
chains = ["TRA", "TRB"]
seqtype = "TCR"

df_metrics = pd.DataFrame(columns=[
            "Index", "UMI_Counts", "UMIs_Mapped_To_Any_VDJ_Gene", "UMIs_with_CDR3", "UMIs_with_Correct_CDR3",
            "UMIs_Mapped_Confidently_To_VJ_Gene"
            ])
for chain in chains:
    df_metrics[f"UMIs_Mapped_To_{chain}"] = ''
```

```python
df_metrics
```

```python
def format_value(value, total):
    if not isinstance(value, numbers.Number):
        return value
    display = str(format(value, ','))
    if total:
        fraction = round(value / total * 100, 2)
        display += f'({fraction}%)'
    return display
```

```python
for index, data in df.groupby("index"):
    
    umi_count = data.shape[0]
    
    data = data[(data["v_call"]!="") | ((data["d_call"]!="")) | ((data["j_call"]!=""))]
    UMIs_Mapped_To_Any_VDJ_Gene = data.shape[0]
    
    df_cdr3 = data[(data["cdr3_aa"]!="") & (data["junction_aa"]!="")]
    UMIs_with_CDR3 = df_cdr3.shape[0]
    
    df_correct_cdr3 = df_cdr3[~(df_cdr3["cdr3_aa"].str.contains(r"\*")) & ~(df_cdr3["cdr3_aa"].str.contains("X"))]
    UMIs_with_Correct_CDR3 = df_correct_cdr3.shape[0]
    
    df_confident = df_correct_cdr3[df_correct_cdr3["productive"]=="T"]
    UMIs_Mapped_Confidently_To_VJ_Gene = df_confident.shape[0]


    UMIs_Mapped_To_chains = []
    for chain in chains:
        df_chain = df_confident[df_confident.locus == chain]
        UMIs_Mapped_To_chain = df_chain.shape[0]
        UMIs_Mapped_To_chains.append(UMIs_Mapped_To_chain)
    
    result = [index, umi_count, UMIs_Mapped_To_Any_VDJ_Gene, UMIs_with_CDR3, UMIs_with_Correct_CDR3, UMIs_Mapped_Confidently_To_VJ_Gene] + UMIs_Mapped_To_chains
    result = [format_value(value, umi_count) for value in result]
    df_metrics.loc[len(df_metrics.index)] = result
    
    print("===========")
    
    
```

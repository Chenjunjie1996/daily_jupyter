```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# BCR
```

```python
df = pd.read_csv("/SGRNJ06/randd/USER/cjj/celedev/vdj_cdr3/mixcr_igblast_compare/igblast_tcr/AD32-T/04.mapping_vdj/AD32-T_airr.tsv", sep='\t')
```

```python
df.fillna("", inplace=True)
```

```python
df = df[(df["cdr3_aa"]!="") & (df["junction_aa"]!="")]
```

```python
df = df[~(df["cdr3_aa"].str.contains(r"\*")) & ~(df["cdr3_aa"].str.contains("X"))]
```

```python
df = df[df["productive"]=='T']
```

```python
df
```

```python
df_cdr3_start = df[["cdr3_end"]]
```

```python
df_cdr3_start
```

```python
min(df_cdr3_start.cdr3_end)
```

```python
max(df_cdr3_start.cdr3_end)
```

```python
df_cdr3_start.cdr3_end.mean()
```

```python
df_cdr3_start.cdr3_end.median()
```

```python
x = ["min", "max", "mean", "median"]
y = [min(df_cdr3_start.cdr3_start), max(df_cdr3_start.cdr3_start), df_cdr3_start.cdr3_start.mean(), df_cdr3_start.cdr3_start.median()]
```

```python
ax = plt.axes()
ax.set_facecolor("white")
ax.grid(False)
plt.bar(x, y, color="steelblue")

for a,b,i in zip(x,y,range(len(x))):
    plt.text(a,b,"%.2f"%y[i], ha="center",fontsize=10)

plt.ylabel("CDR3_start")
plt.title("TCR")
plt.savefig("/SGRNJ06/randd/USER/cjj/celedev/vdj_cdr3/mixcr_igblast_compare/tcr.png", dpi=300, bbox_inches="tight")
```

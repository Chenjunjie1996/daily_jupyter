```python
airr = pd.read_csv("/SGRNJ06/randd/USER/cjj/celedev/vdj10x/20230423hsV_ratC/N1001-B/03.assemble/N1001-B/outs/airr_rearrangement.tsv", sep='\t')
```

```python
df = pd.read_csv("/SGRNJ06/randd/USER/cjj/celedev/vdj10x/20230423hsV_ratC/N1001-B/03.assemble/N1001-B/outs/filtered_contig_annotations.csv")
```

```python
bc_chain_dict = df.groupby("barcode")["chain"].apply(set).to_dict()
```

```python
pair_chain_dict = {key: value for key, value in bc_chain_dict.items() if len(value)>=2 and "IGH" in value}
```

```python
df = df[df["barcode"].isin(pair_chain_dict)]
```

```python
df_pair_chain = df.groupby("barcode").filter(lambda x: (len(x) == 2))
```

```python
df_pair_chain
```

```python
df_multi_chain = df.groupby("barcode").filter(lambda x: (len(x) > 2))
```

```python
len(set(df_multi_chain.barcode))
```

```python
airr = airr[airr["cell_id"].isin(set(df_pair_chain.barcode))]
```

```python
airr
```

```python
df_pair_chain = df_pair_chain.rename(columns={"contig_id":"sequence_id"})
```

```python
df_pair_chain
```

```python
airr = pd.merge(airr, df_pair_chain ,how='inner', on='sequence_id')
```

```python
airr["cell_id"] = airr["cell_id"].apply(lambda x: tenX_sgr[x.split('-')[0]])
airr["sequence_id"] = airr["sequence_id"].apply(lambda x: tenX_sgr[x.split('-')[0]]+'_'+x.split('_')[1]+'_'+x.split('_')[2])
```

```python
airr["chain"] = airr["j_call"].apply(lambda x: x[0:3])
```

```python
airr
```

```python
airr = airr[["cell_id","clone_id","sequence_id","chain", "sequence","sequence_aa", "reads"]]
```

```python
airr["temp"] = airr["clone_id"].apply(lambda x: int(x[9:]))
airr = airr.sort_values(["temp","cell_id", "chain"])
del airr["temp"]
```

```python
df0 = pd.read_csv("/SGRNJ06/randd/USER/cjj/celedev/vdj10x/20230309ningxiaBCR_downsample/5g/SS_02/03.assemble/SS_02/outs/airr_rearrangement.tsv", sep='\t')
df1 = pd.read_csv("/SGRNJ06/randd/USER/cjj/celedev/vdj10x/20230301BCR/Hum_0202PBMC_SLGOT_1_B3lib/03.assemble/Hum_0202PBMC_SLGOT_1_B3lib/outs/airr_rearrangement.tsv", sep='\t')
df2 = pd.read_csv("/SGRNJ06/randd/USER/cjj/celedev/vdj10x/20230301BCR/Hum_0202PBMC_SLGOT_DE_2_B3lib/03.assemble/Hum_0202PBMC_SLGOT_DE_2_B3lib/outs/airr_rearrangement.tsv", sep='\t')
df = pd.concat([df0,df1,df2])
```

```python
df
```

```python
df_c_start_end = df[["c_sequence_start", "c_sequence_end"]]
```

```python
df_c_start_end["c_sequence_length"] = df_c_start_end["c_sequence_end"] - df_c_start_end["c_sequence_start"]
```

```python
df_c_start_end.c_sequence_length.mean()
```

```python
df_c_start_end.c_sequence_length.median()
```

```python
x = ["min", "max", "mean", "median"]
y = [min(df_c_start_end.c_sequence_length), max(df_c_start_end.c_sequence_length), df_c_start_end.c_sequence_length.mean(), df_c_start_end.c_sequence_length.median()]
```

```python
ax = plt.axes()
ax.set_facecolor("white")
ax.grid(False)
plt.bar(x, y, color="steelblue")

for a,b,i in zip(x,y,range(len(x))):
    plt.text(a,b,"%.2f"%y[i], ha="center",fontsize=10)

plt.ylabel("C_region_length")
plt.title("BCR")
plt.savefig("/SGRNJ06/randd/USER/cjj/celedev/vdj10x/20230323Cregion/bcr.png", dpi=300, bbox_inches="tight")
```

```python
df = df.sort_values("barcode")
```

```python
df['chain_cdr3aa'] = df[['chain', 'cdr3']].apply(':'.join, axis=1)
```

```python
df = df.groupby("barcode").filter(lambda x: (len(x) >= 2))
```

```python
grouped = df.groupby(["chain_cdr3aa"])
```

```python
res = set()
for name, data in grouped:
    if len(set(data.raw_clonotype_id))>=2:
        #print(name)
        #print(data.raw_clonotype_id)
        res.update(set(data.raw_clonotype_id))
```
